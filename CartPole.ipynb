```python
import tensorflow as tf
from Strategy import RL
```

```python
case_num = 3

# 1 : epsilon , 2 : softmax , 3 : sparsemax , 4 : sparse+eps
Exp_list = ['epsilon', 'epsilon', 'sparse+eps']

Double_list = [True, True, False]
Dueling_list = [True, True, True]
Prioritized_list = [True, True, True]

for i in range(case_num):
    tf.reset_default_graph()
    with tf.Session() as sess :
        CartPole = RL(sess, dis = 0.99, REPLAY_MEMORY = 10000, batch_size = 16, max_episodes = 500,
                      training_step = 30, copy_step = 30, action_res = None,
                      layer_size = 128, learning_rate = 0.00001, repu_num = 5,
                      alpha = 0.8, beta_init = 0.4, eps = 0.01, eps_div = 2, save_epi = 50,
                      ending_cond_epis = 100, ending_cond_reward = 195, reward_p = 100, reward_n = -100,
                      seed_n = i, Game = 'CartPole-v0')
    
        CartPole.run_RL(case_n = i+1, Exp = Exp_list[i], seed_n = i,
                       Double = Double_list[i], Dueling = Dueling_list[i], Prioritized = Prioritized_list[i])
    sess.close()
```
