```python
import pickle
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np


##########################################################################################

EP1 = ['Hopper_E(15)_(16,1,4,4,1,0.00005)', 'EP :: BS : 16']
EP2 = ['Hopper_E(15)_(32,1,4,4,1,0.00005)', 'EP ::  AC : 3375, BS : 32']
EP3 = ['Hopper_E(15)_(64,1,4,4,1,0.00005)', 'EP :: BS : 64']

EP4 = ['Hopper_E(9)_(32,1,4,4,1,0.00005)', 'EP ::  AC : 729,   BS : 32']
EP5 = ['Hopper_E(21)_(32,1,4,4,1,0.00005)', 'EP :: AC : 21, BS : 32']

##########################################################################################

SO1 = ['Hopper_S(15)_(16,1,4,4,0.5,1,0.00005)', 'SO :: AC : 3375, BS : 16, SCALE : 0.5']
SO2 = ['Hopper_S(15)_(16,1,4,4,0.2,1,0.00005)', 'SO :: BS : 16, SCALE : 0.2']
SO3 = ['Hopper_S(15)_(16,1,4,4,0.1,1,0.00005)', 'SO :: BS : 16, SCALE : 0.1']

SO4 = ['Hopper_S(15)_(32,1,4,4,0.5,1,0.00005)', 'SO :: AC : 3375, BS : 32, SCALE : 0.5']
SO5 = ['Hopper_S(15)_(32,1,4,4,0.2,1,0.00005)', 'SO :: BS : 32, SCALE : 0.2']
SO6 = ['Hopper_S(15)_(32,1,4,4,0.1,1,0.00005)', 'SO :: BS : 32, SCALE : 0.1']

SO7 = ['Hopper_S(15)_(64,1,4,4,0.5,1,0.00005)', 'SO :: BS : 64, SCALE : 0.5']
SO8 = ['Hopper_S(15)_(64,1,4,4,0.2,1,0.00005)', 'SO :: BS : 64, SCALE : 0.2']
SO9 = ['Hopper_S(15)_(64,1,4,4,0.1,1,0.00005)', 'SO :: BS : 64, SCALE : 0.1']

SO10 = ['Hopper_S(15)_(32,1,4,4,1,1,0.00005)', 'SO :: BS : 32, SCALE : 1']
SO11 = ['Hopper_S(15)_(32,1,4,4,0.01,1,0.00005)', 'SO :: BS : 32, SCALE : 0.01']

SO12 = ['Hopper_S(9)_(32,1,4,4,0.5,1,0.00005)', 'SO :: AC : 9, BS : 32, SCALE : 0.5']
SO13 = ['Hopper_S(9)_(32,1,4,4,0.2,1,0.00005)', 'SO :: AC : 9, BS : 32, SCALE : 0.2']
SO14 = ['Hopper_S(9)_(32,1,4,4,0.1,1,0.00005)', 'SO :: AC : 9, BS : 32, SCALE : 0.1']

SO15 = ['Hopper_S(21)_(32,1,4,4,0.5,1,0.00005)', 'SO :: AC : 21, BS : 32, SCALE : 0.5']
SO16 = ['Hopper_S(21)_(32,1,4,4,0.2,1,0.00005)', 'SO :: AC : 21, BS : 32, SCALE : 0.2']
SO17 = ['Hopper_S(21)_(32,1,4,4,0.1,1,0.00005)', 'SO :: AC : 21, BS : 32, SCALE : 0.1']

##########################################################################################

SP1 = ['Hopper_SP(15)_(16,1,4,4,0.1,1,0.00005)', 'SP :: BS : 16, SCALE : 0.1']
SP2 = ['Hopper_SP(15)_(16,1,4,4,0.05,1,0.00005)', 'SP :: BS : 16, SCALE : 0.05']
SP3 = ['Hopper_SP(15)_(16,1,4,4,0.01,1,0.00005)', 'SP :: BS : 16, SCALE : 0.01']

SP4 = ['Hopper_SP(15)_(32,1,4,4,0.1,1,0.00005)', 'SP :: BS : 32, SCALE : 0.1']
SP5 = ['Hopper_SP(15)_(32,1,4,4,0.05,1,0.00005)', 'SP :: BS : 32, SCALE : 0.05']
SP6 = ['Hopper_SP(15)_(32,1,4,4,0.01,1,0.00005)', 'SP ::  AC : 3375, BS : 32, SCALE : 0.01']

SP7 = ['Hopper_SP(15)_(64,1,4,4,0.1,1,0.00005)', 'SP :: BS : 64, SCALE : 0.1']
SP8 = ['Hopper_SP(15)_(64,1,4,4,0.05,1,0.00005)', 'SP :: BS : 64, SCALE : 0.05']
SP9 = ['Hopper_SP(15)_(64,1,4,4,0.01,1,0.00005)', 'SP :: BS : 64, SCALE : 0.01']

SP10 = ['Hopper_SP(15)_(32,1,4,4,0.5,1,0.00005)', 'SP :: BS : 32, SCALE : 0.5']
SP11 = ['Hopper_SP(15)_(32,1,4,4,0.001,1,0.00005)', 'SP :: BS : 32, SCALE : 0.001']

SP12 = ['Hopper_SP(9)_(32,1,4,4,0.1,1,0.00005)', 'SP :: AC : 9, BS : 32, SCALE : 0.1']
SP13 = ['Hopper_SP(9)_(32,1,4,4,0.05,1,0.00005)', 'SP :: AC : 9, BS : 32, SCALE : 0.05']
SP14 = ['Hopper_SP(9)_(32,1,4,4,0.01,1,0.00005)', 'SP :: AC : 9, BS : 32, SCALE : 0.01']

SP15 = ['Hopper_SP(21)_(32,1,4,4,0.1,1,0.00005)', 'SP :: AC : 21, BS : 32, SCALE : 0.1']
SP16 = ['Hopper_SP(21)_(32,1,4,4,0.05,1,0.00005)', 'SP :: AC : 21, BS : 32, SCALE : 0.05']
SP17 = ['Hopper_SP(21)_(32,1,4,4,0.01,1,0.00005)', 'SP :: AC : 21, BS : 32, SCALE : 0.01']

##########################################################################################

iteration = 3
steps = 1500000

file_name_list = [EP2, SO1, SP6]

#file_name_list = [SO1, SO2, SO3]
#file_name_list = [SP1, SP2, SP3]
#file_name_list = [EP1, SO1, SP2]

#file_name_list = [SO4, SO5, SO6, SO10, SO11]
#file_name_list = [SP4, SP5, SP6, SP10, SP11]
#file_name_list = [EP2, SO4, SP6]

#file_name_list = [SO7, SO8, SO9]
#file_name_list = [SP7, SP8, SP9]
#file_name_list = [EP3, SO7, SP9]

#file_name_list = [SO12, SO13, SO14]
#file_name_list = [SP12, SP13, SP14]
#file_name_list = [EP4, SO12, SP14]

#file_name_list = [SO15, SO16, SO17]
#file_name_list = [SP15, SP16, SP17]
#file_name_list = [EP5, SO15, SP15]

file_name_list = [EP4, SO4, SP6]

graph_color = ['r', 'g', 'b', 'y', 'm', 'c']

####################################
max_list = []

max_avg_list = []
max_avg_epi_list = []
std_list = []

for i in range(len(file_name_list)):
    step_avg_list = []
    global_step_list = []
    temp_list = []

    for j in range(iteration):
        with open('/home/jolp/Desktop/Data/'+file_name_list[i][0]+'_seed'+str(j+1), 'rb') as fin1:
            step_avg_list.append(pickle.load(fin1))
        
        with open('/home/jolp/Desktop/Data/'+file_name_list[i][0]+'_global__seed'+str(j+1), 'rb') as fin2:
            global_step_list.append(pickle.load(fin2))
            
        temp_list.append(global_step_list[j][-1])
    
    step_list = []
    for it in range(iteration):
        step = 0
        step_list_temp = []
        for num in range(steps):
            if num == global_step_list[it][step]:
                step += 1
            step_list_temp.append(step_avg_list[it][step])
        step_list.append(step_list_temp)
    
    step_list = np.array(step_list)
    
    print_step_avg_list = np.average(step_list, axis=0)
    print_step_std_list = np.std(step_list, axis=0)
    
    max_list.append(np.max(step_list, axis=1))
    
    max_avg_list.append(np.max(print_step_avg_list))
    max_avg_epi_list.append(np.argmax(print_step_avg_list))
    std_list.append(np.average(print_step_std_list))
    
    len_file = len(print_step_avg_list)
    x_values = list(range(1, len_file+1))
    y_values = print_step_avg_list[:]
    
    plt.plot(x_values, y_values, c=graph_color[i], label=file_name_list[i][1])
    
    gap = 250
    x_s_values = [gap*o for o in range(1, len_file/gap+1)]
    y_a_values = [print_step_avg_list[x_s_values[o]-gap] for o in range(len_file/gap)]
    y_s_values = [print_step_std_list[x_s_values[o]-gap] for o in range(len_file/gap)]
    y_s_values = np.array(y_s_values)
    
    #plt.errorbar(x_values, y_values, yerr = 0.5*print_step_std_list, ecolor=graph_color[i], alpha=0.01)
    plt.errorbar(x_s_values, y_a_values, yerr = 0.5*y_s_values, ecolor=graph_color[i], alpha=0.01)


plt.title('Hopper, Exploration')
plt.grid(True)
plt.legend(loc=2, bbox_to_anchor=(1., 0.3))
plt.show()

print('################################################')
for i in range(len(file_name_list)):
    print('')
    print('GRAPH{} : {}'.format(i+1,file_name_list[i][1]))
    print('    MAX : {}, STD : {}'.format(round(max_avg_list[i],3), round(std_list[i],3)))
    if iteration == 3:
        print('    MAX1 :: {}, MAX2 :: {}, MAX3 :: {}'.format(round(max_list[i][0],3),round(max_list[i][1],3),round(max_list[i][2],3)))
    if iteration == 2:
        print('    MAX1 :: {}, MAX2 :: {}'.format(round(max_list[i][0],3),round(max_list[i][1],3)))
print('')
print('################################################')
```
